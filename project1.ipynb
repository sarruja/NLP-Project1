Kurzer Check ob test.csv file als Test verwended wird oder train.csv in train/test gesplitet wird.
Fazit: Die test_labels.csv von Kaggle enthält für viele Einträge den Label -1 – das bedeutet: "dieser Kommentar wurde nicht bewertet" und soll beim Scoring ignoriert werden. Nur ein Teil hat echte Labels (0 oder 1).

```python
# Test labels
df_test_labels = pd.read_csv('data/test_labels.csv')
print("=== test_labels.csv ===")
print("Shape:", df_test_labels.shape)
print(df_test_labels['toxic'].value_counts())
print(f"Anteil -1: {round(100 * (df_test_labels['toxic'] == -1).sum() / len(df_test_labels), 1)}%")

# Test data
df_test = pd.read_csv('data/test.csv')
print("\n=== test.csv ===")
print("Shape:", df_test.shape)
print(df_test.head(2))

# Train data
df_train = pd.read_csv('train.csv', encoding='utf-8')
print("\n=== train.csv ===")
print("Shape:", df_train.shape)
print("Label-Verteilung:")
print(df_train['toxic'].value_counts())
print(f"Anteil toxic: {round(100 * df_train['toxic'].sum() / len(df_train), 1)}%")
```
